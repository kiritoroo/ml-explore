```python
from tensorflow import keras
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
MODEL_PATH = "./models/model_animal.h5"
ANIMALS_CATEGORIES = ["Cat", "Cow", "Dog", "Elephant", "Panda"]

model_animal = keras.models.load_model(MODEL_PATH)
```

```python
def debug_imshow(image):
  fig = plt.figure(figsize=(4, 3), dpi=100)
  plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
  plt.show()
```

```python
image_path = 'static/test1.webp'
image_input = cv.imread(image_path)
debug_imshow(image_input)
cv.imwrite("out.jpg", image_input)
```

```python
WIDTH_INPUT = 256
HEIGHT_INPUT = 256
dim = (WIDTH_INPUT, HEIGHT_INPUT)
image_input = cv.resize(image_input, dim)
image_input= np.expand_dims(image_input, axis=0)
image_input = np.array(image_input)
image_input = image_input / 255

prediction = model_animal.predict(image_input)

print(ANIMALS_CATEGORIES[np.argmax(prediction)])
```
